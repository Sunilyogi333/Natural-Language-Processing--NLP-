```python
import pandas as pd
data={
    'text':['I like playing football','I like swimming', 'I dont like walking', 'I like watching football'],
    'label':['positive','positive','negative','positive']
}
df = pd.DataFrame(data)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])

print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["i like sleeping and eating"]).toarray()
```

```python
cv2 = CountVectorizer(ngram_range=(2,2)) #bigram nodel
bow2 = cv2.fit_transform(df['text'])

cv2.vocabulary_
```

```python
cv3 = CountVectorizer(ngram_range=(1,3)) #uni, bi, trigram
bow3 = cv3.fit_transform(df['text']).toarray()
cv3.vocabulary_
#
```
